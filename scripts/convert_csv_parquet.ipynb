```python
import polars as pl
from pathlib import Path
from tqdm import tqdm

# Pfad zur CSV-Datei (relativ zum Notebook-Standort)
csv_path = Path(__file__).parent.parent / 'data' / 'raw' / 'speeches.csv' if '__file__' in globals() else Path('../data/raw/speeches.csv')

print("ðŸ”„ Starte Konvertierung...")

# CSV laden
with tqdm(total=3, desc="Fortschritt", bar_format='{l_bar}{bar}| {n_fmt}/{total_fmt}') as pbar:
    pbar.set_description("ðŸ“– Lade CSV-Datei")
    df = pl.read_csv(csv_path)
    pbar.update(1)
    
    # Als Parquet speichern
    pbar.set_description("ðŸ’¾ Schreibe Parquet-Datei")
    parquet_path = csv_path.with_suffix('.parquet')
    df.write_parquet(parquet_path)
    pbar.update(1)
    
    pbar.set_description("âœ… Abgeschlossen")
    pbar.update(1)

print(f"\nâœ… CSV erfolgreich konvertiert und gespeichert als: {parquet_path}")
print(f"ðŸ“Š Anzahl der Zeilen: {len(df):,}")
print(f"ðŸ“‹ Anzahl der Spalten: {len(df.columns)}")
```
